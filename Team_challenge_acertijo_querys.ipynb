```python
import pandas as pd
import sqlite3

connection = sqlite3.connect("data/sql-murder-mystery.db")


cursor_bootcamp = connection.cursor()

def sql_query(query):

   
    cursor_bootcamp.execute(query)
    

    
    ans = cursor_bootcamp.fetchall()

    
    names = [description[0] for description in cursor_bootcamp.description]

    return pd.DataFrame(ans,columns=names)
```

```python
query = "select * from crime_scene_report where date = 20180115 AND type = 'murder'"#Primero vemos todos los asesinatos que coincidan con la fecha del asesinato
sql_query(query)
```

```python
query = "select * from interview where person_id = 16371" #Lo pasamos por la tabla entrevista 
sql_query(query)
```
